```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


x1 = np.random.normal(0, 1, 20)
y1 = np.random.normal(0, 1, 20)
x2 = np.random.normal(0, 1, 20)
y2 = np.random.normal(0, 1, 20)

plt.scatter(x1, y1, c='green')
plt.scatter(x2, y2, c='red')
plt.show()

df = pd.DataFrame({'x': np.concatenate([x1, x2]), 'y': np.concatenate([y1, y2]), 
                   'label': np.concatenate([np.zeros(20, dtype=int), np.ones(20, dtype=int)])})
df.to_csv('../data/input_space.csv', index=False)
```
