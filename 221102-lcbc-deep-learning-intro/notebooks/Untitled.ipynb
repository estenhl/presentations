# Building a neural network: Setup

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 6)

x
```

# Building a neural network: Linear regression

```python
a = -0.5
b = -2

def f(x):
    return (a * x) + b

y = f(x)

plt.plot(x, y)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
```

# Building a neural network: Artificial neural network

```python
w0 = 0.5
w1 = -0.75
w2 = 0.25
b0 = -1
b1 = -0.25
b2 = 0.75

def f(x):
    return np.maximum(0, (w0 * x) + b0) + \
           np.maximum(0, (w1 * x) + b1) + \
           np.maximum(0, (w2 * x) + b2)

y = f(x)

plt.plot(x, y)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
```
