```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.random.normal(27, 4, 30)
y = 25 + 4.5 * x + np.random.normal(0, 5, 30)

plt.scatter(x, y)
plt.show()

for i in range(len(x)):
    print(f'({x[i]:.2f},{y[i]:.2f})')
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

np.random.seed(4)

x = np.random.normal(0, 4, 30)
y = sigmoid(x)#6 * x + np.random.normal(0, 5, 20)

x = x + 25
y = y * 60 + 110 + np.random.normal(0, 3, 30)

plt.scatter(x, y)
plt.show()

for i in range(len(x)):
    print(f'({x[i]:.2f},{y[i]:.2f})')
```
