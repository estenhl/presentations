```python
import pandas as pd


df = pd.read_csv('/Users/esten/Downloads/Auto.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler

df[['horsepower', 'weight']] = StandardScaler().fit_transform(df[['horsepower', 'weight']])
```

```python
from sklearn.linear_model import Ridge

for alpha in [1, 1000, 100000]:
    model = Ridge(alpha=alpha)
    model.fit(df[['horsepower', 'weight']], df['mpg'])
    print(model.coef_)
```
