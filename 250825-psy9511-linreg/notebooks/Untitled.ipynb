```python
import numpy as np
import matplotlib.pyplot as plt

# Create a vector of 100 standard normally distributed numbers
data = np.random.randn(100)

# Create a histogram of the data
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of 100 Standard Normally Distributed Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
