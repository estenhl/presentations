### Read in the data frame

```python
import pandas as pd


df = pd.read_csv('~/Downloads/Auto.csv')
df
```

### Fit a linear regression model

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(df[['horsepower']], df['mpg'])
```

### Plot the predictions of the model

```python
import matplotlib.pyplot as plt

predictions = model.predict(df[['horsepower']])
plt.scatter(df['mpg'], predictions)
```

### Print the intercept and coefficients

```python
print(model.intercept_)
print(model.coef_)
```

### Fit a multivariate linear regression model

```python
multivariate = LinearRegression()
multivariate.fit(df[['horsepower', 'weight']], df['mpg'])
print(multivariate.intercept_)
print(multivariate.coef_)
```

### Demonstrate k-nearest neighbours

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

for model in [LinearRegression(), KNeighborsRegressor(3), KNeighborsRegressor(5)]:
    model.fit(df[['horsepower']], df['mpg'])
    predictions = model.predict(df[['horsepower']])
    mse = mean_squared_error(df['mpg'], predictions)
    print(mse)
```
