```python
import numpy as np
import matplotlib.pyplot as plt


d1 = np.random.normal(0, 0.3, (20, 2))
d2 = np.random.normal(0.5, 0.3, (20, 2))

plt.scatter(d1[:,0], d1[:,1], c='red')
plt.scatter(d2[:,0], d2[:,1], c='blue')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame(np.concatenate([d1, d2]), columns=['x1', 'x2',])
df['y'] = np.concatenate([np.zeros(20), np.ones(20)])
df.to_csv('../data/bayesdata.csv', index=False)
print(df.sample(frac=1.).head())
```
