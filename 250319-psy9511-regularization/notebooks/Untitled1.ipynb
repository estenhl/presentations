```python
import pandas as pd

df = pd.read_csv('~/Downloads/Credit.csv')

df
```

```python
df = pd.get_dummies(df, columns=['Region'])

df.head()
```

```python
for col in df.columns:
    if col.startswith('Region'):
        df[col] = df[col].astype(int)

df
df
```

```python
import numpy as np

predictors = ['Age', 'Education', 'Region_East', 'Region_South', 'Region_West']
target = 'Balance'


df[target] = df[target].apply(lambda x: x > np.median(df[target]).astype(int))


df = df.sort_values([target, 'Age'])
df['fold'] = np.arange(len(df)) % 5
df
```

```python
df['Balance'] = df['Balance'].astype(int)

train = df[df['fold'] != 2]
validation = df[df['fold'] == 2]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train[predictors], train[target])

predictions = model.predict_proba(train[predictors])
```

```python
predictions = predictions[:,1]
predictions
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report

print(accuracy_score(train[target], np.round(predictions)))
print(confusion_matrix(train[target], np.round(predictions)))
print(classification_report(train[target], np.round(predictions)))
```
