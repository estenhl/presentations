```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


df = pd.read_csv('~/Downloads/Heart.csv')

predictors = ['RestBP', 'ChestPain']
target = 'Thal'

df = df[predictors + [target]]
df = pd.get_dummies(df, columns=['ChestPain'])
df[target] = df[target].apply(lambda x: 0 if x =='normal' else 1)

train_X = df.drop(columns=[target]).values
train_X
train_y = df[target].values

model = LinearDiscriminantAnalysis()
model.fit(train_X, train_y)
predictions = model.predict(train_X)

from sklearn.metrics import accuracy_score
import numpy as np

print(accuracy_score(train_y, np.round(predictions)))
```
