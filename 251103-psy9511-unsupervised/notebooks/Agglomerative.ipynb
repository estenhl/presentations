```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


np.random.seed(42)

c1 = np.random.normal(-0.05, 0.05, (5, 2))
c2 = np.random.normal(0.15, 0.05, (5, 2))
c3 = np.column_stack([
    np.random.normal(0.15, 0.05, 5),
    np.random.normal(-0.1, 0.05, 5)
])

plt.scatter(c1[:,0], c1[:,1], c='blue')
plt.scatter(c2[:,0], c2[:,1], c='blue')
plt.scatter(c3[:,0], c3[:,1], c='blue')
plt.show()

x = np.concatenate([c1[:,0], c2[:,0], c3[:,0]])
y = np.concatenate([c1[:,1], c2[:,1], c3[:,1]])

data = np.column_stack([x, y])

print(data)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```
